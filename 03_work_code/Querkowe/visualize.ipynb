### Feature Scaling

```python
scale_dict = joblib.load('scale.pkl')
# scale_dict
```

```python
scale_summary_dict = {}

scale_keys = ['original', 'minmax', 'standard', 'robust']

for scale_key in scale_keys:
  scale_summary_dict[scale_key] = {}

  scale_summary_dict[scale_key]['catboost tuned'] = scale_dict[scale_key]['catboost tuned']
  scale_summary_dict[scale_key]['blend5'] = scale_dict[scale_key]['blend5']
  scale_summary_dict[scale_key]['blend12'] = scale_dict[scale_key]['blend12']
  del scale_dict[scale_key]['catboost tuned']
  del scale_dict[scale_key]['blend5']
  del scale_dict[scale_key]['blend12']
```
